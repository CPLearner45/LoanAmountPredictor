Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading csv file

```python
df = pd.read_csv("LoanAmountPrediction.csv")
```

Checking null and duplicate values

```python
df.shape
```

```python
df.isnull().sum() * 100 / len(df)
```

```python
df.info()
```

```python
#dropping id coloumn
df = df.drop('Loan_ID',axis = 1)
```

```python
df
```

```python
columns = ['Gender','Dependents','LoanAmount','Loan_Amount_Term']
df = df.dropna(subset=columns)
df.isnull().sum()
```

```python
df['Self_Employed'] =df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Credit_History'] =df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df.isnull().sum()
```

Handling categorical df

```python
df['Dependents'] = df['Dependents'].replace(to_replace ="3+", value = "4")
df['Dependents'].unique()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0}).astype('int')
df['Married'] = df['Married'].map({'Yes':1,'No':0}).astype('int')
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0}).astype('int')
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1,'No':0}).astype('int')
df['Property_Area'] = df['Property_Area'].map({'Rural':0,'Semiurban':2,'Urban':1}).astype('int')
df['Loan_Status'] = df['Loan_Status'].map({'Y':1,'N':0}).astype('int')

df.sample(10)
```

Splitting data into target and vector

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

Feature scaling

```python
columns = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X[columns]=st.fit_transform(X[columns])
```

```python
X.sample(5)
```

Splitting training and testing data sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Linear regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

regressor = LogisticRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

print(f"{regressor} accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,X,y,cv=5)
print(f"{regressor} Avg cross val score is {np.mean(cvscore)*100}%")
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

regressor = RandomForestClassifier()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

print(f"{regressor} accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,X,y,cv=5)
print(f"{regressor} Avg cross val score is {np.mean(cvscore)*100}%")
```

Decision tree regressor

```python
from sklearn.tree import DecisionTreeClassifier

regressor = DecisionTreeClassifier()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

print(f"{regressor} accuracy is {accuracy_score(y_test,y_pred)*100}%")

cvscore = cross_val_score(regressor,X,y,cv=5)
print(f"{regressor} Avg cross val score is {np.mean(cvscore)*100}%")
```
